# Visualization

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=df['Car Condition'].value_counts().index, y=df['Car Condition'].value_counts().values)

plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=df['Weather'].value_counts().index, y=df['Weather'].value_counts().values)

plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x=df['Traffic Condition'].value_counts().index, y=df['Traffic Condition'].value_counts().values)

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['distance'], bins=100, kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="distance", y="fare_amount", data=df, alpha=0.4)
plt.title("Fare Amount vs Trip Distance")
plt.xlabel("Trip Distance (km)")
plt.ylabel("Fare ($)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Car Condition", y="fare_amount", data=df, palette="Set3")
plt.title("Fare Amount vs Car Condition")
plt.ylabel("Fare ($)")
plt.xlabel("Car Condition")
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(x="Weather", y="fare_amount", data=df, palette="Set2")
plt.title("Fare Amount by Weather")
plt.ylabel("Fare ($)")
plt.xlabel("Weather")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(x="Traffic Condition", y="fare_amount", data=df, palette="Set3")
plt.title("Fare Amount by Traffic Condition")
plt.ylabel("Fare ($)")
plt.xlabel("Traffic Condition")
plt.xticks(rotation=25)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Weather", y="fare_amount",
            hue="Traffic Condition", data=df,
            palette="pastel")
plt.title("Fare Amount by Weather and Traffic")
plt.ylabel("Fare ($)")
plt.xlabel("Weather")
plt.legend(title="Traffic")
plt.show()
```

```python
# Calculate average fare per month
avg_fare_month = df.groupby("month")["fare_amount"].mean().reset_index()
sns.lineplot(data=avg_fare_month, x="month", y="fare_amount", marker="o", color="blue")
plt.title("Average Fare Amount by Month")
plt.xlabel("Month")
plt.ylabel("Average Fare Amount")
plt.show()
```

```python
yearly_fare = df.groupby("year")["fare_amount"].mean().reset_index()

# Set style
sns.set_style("whitegrid")

# Create lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_fare, x="year", y="fare_amount", marker="o", linewidth=2)

# Add labels and title
plt.title("Average Fare Amount per Year", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Fare ($)", fontsize=14)

# Show grid + plot
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# Avoid divide-by-zero issues
df = df[df["distance"] > 0]
df["fare_per_km"] = df["fare_amount"] / df["distance"]

plt.figure(figsize=(10, 6))
sns.kdeplot(df["fare_per_km"], fill=False, color="teal", lw=2)

plt.title("Fare per Distance Ratio Distribution (Line Plot)")
plt.xlabel("Fare per km ($/km)")
plt.ylabel("Density")
plt.xlim(0, 20)  # zoom in to typical range
plt.show()
```

```python
# Group by hour and traffic condition
traffic_hour = df.groupby(["hour", "Traffic Condition"])["fare_amount"].count().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=traffic_hour,
    x="hour",
    y="fare_amount",
    hue="Traffic Condition",
    marker="o"
)
plt.title("Taxi Demand by Hour of Day & Traffic Condition")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Rides")
plt.xticks(range(0, 24))
plt.legend(title="Traffic Condition")
plt.show()
```

```python
# Group by hour and traffic condition
traffic_hour = df.groupby(["day", "Traffic Condition"])["fare_amount"].count().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=traffic_hour,
    x="day",
    y="fare_amount",
    hue="Traffic Condition",
    marker="o"
)
plt.title("Taxi Demand by Day of Month & Traffic Condition")
plt.xlabel("Day of Month")
plt.ylabel("Number of Rides")
plt.xticks(rotation=45)
plt.legend(title="Traffic Condition")
plt.show()
```

```python
# Group by hour and traffic condition
traffic_hour = df.groupby(["weekday", "Traffic Condition"])["fare_amount"].count().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=traffic_hour,
    x="weekday",
    y="fare_amount",
    hue="Traffic Condition",
    marker="o"
)
plt.title("Taxi Demand by Day of Week & Traffic Condition")
plt.xlabel("Day of Week")
plt.ylabel("Number of Rides")
plt.xticks(rotation=45)
plt.legend(title="Traffic Condition")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dist_cols = ['jfk_dist','ewr_dist','lga_dist','sol_dist','nyc_dist']

for c in dist_cols:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df[c], y=df['fare_amount'], alpha=0.3)
    plt.title(f"Fare vs {c}")
    plt.xlabel(c)
    plt.ylabel("Fare Amount ($)")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="passenger_count", y="fare_amount", data=df, alpha=0.4)
plt.title("Fare Amount vs Passengers")
plt.xlabel("Passengers_count")
plt.ylabel("Fare ($)")
plt.show()
```

```python
# Group by weekday and hour, then compute average fare
fare_heatmap = df.groupby(["weekday", "hour"])["fare_amount"].mean().reset_index()

# Pivot for heatmap format
fare_pivot = fare_heatmap.pivot(index="weekday", columns="hour", values="fare_amount")

# Plot heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(fare_pivot, cmap="YlOrRd", annot=False, cbar_kws={'label': 'Average Fare ($)'})

# Customize ticks
plt.title("Average Fare Heatmap by Weekday vs Hour", fontsize=16)
plt.xlabel("Hour of Day", fontsize=14)
plt.ylabel("Weekday (0=Mon, 6=Sun)", fontsize=14)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average fare per hour
avg_fare_hour = df.groupby("hour")["fare_amount"].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(
    data=avg_fare_hour,
    x="hour",
    y="fare_amount",
    palette="coolwarm"
)
plt.title("Average Fare Amount vs Hour of the Day")
plt.xlabel("Hour of Day")
plt.ylabel("Average Fare Amount")
plt.xticks(rotation=0)
plt.show()
```

```python
# Count trips for each Traffic Condition by hour
traffic_hour = df.groupby(['hour', 'Traffic Condition']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.barplot(
    data=traffic_hour,
    x='hour',
    y='count',
    hue='Traffic Condition',
    palette='Set2'
)
plt.title("Traffic Condition vs Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Trips")
plt.legend(title="Traffic Condition")
plt.show()
```

```python
# Count trips for each Traffic Condition by weekday
traffic_weekday = df.groupby(['weekday', 'Traffic Condition']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.barplot(
    data=traffic_weekday,
    x='weekday',
    y='count',
    hue='Traffic Condition',
    palette='Set2'
)
plt.title("Traffic Condition vs Day of Week")
plt.xlabel("Day of Week (0=Monday)")
plt.ylabel("Number of Trips")
plt.legend(title="Traffic Condition")
plt.show()
```

```python
# Count trips for each Traffic Condition by weekday
traffic_weekday = df.groupby(['month', 'Traffic Condition']).size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.barplot(
    data=traffic_weekday,
    x='month',
    y='count',
    hue='Traffic Condition',
    palette='Set2'
)
plt.title("Traffic Condition vs month")
plt.xlabel("Months")
plt.ylabel("Number of Trips")
plt.legend(title="Traffic Condition")
plt.show()
```

```python
# Count trips per day of the month
trips_per_day = df['weekday'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
sns.barplot(x=trips_per_day.index, y=trips_per_day.values, palette="viridis")
plt.title("Number of Trips per Weekday")
plt.xlabel("Day of Week")
plt.ylabel("Number of Trips")
plt.show()

# Find day with most trips
most_trips_day = trips_per_day.idxmax()
print(f"Day with most trips: {most_trips_day} ({trips_per_day.max()} trips)")
```
